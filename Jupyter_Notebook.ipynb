### This code implements a function named fibonacci that calculates the values in the Fibonacci sequence up to a specified number of elements. The Fibonacci sequence is a mathematical sequence in which each number is the sum of the two preceding ones, starting from 0 and 1. Fn = Fn-1 + Fn-2 which F0 = 0 และ F1 = 1


```python
def fibonacci(n):
    # Base case 1: If n is 0, return 0
    if n == 0:
        return 0
    # Base case 2: If n is 1, return 1
    elif n == 1:
        return 1
    # Recursive case: If n is greater than 1, return the sum of the previous two values in the sequence
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

if __name__ == "__main__":
    length = int(input("Enter the length of the fibonacci sequence: "))
    for i in range(length):
        print(fibonacci(i), end=", ")
```
